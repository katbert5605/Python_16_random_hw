Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
\**7: 3 руб.
\*00: 2 руб.
\**0: 1 руб.

### Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

\* – любая цифра

```python
# Попробуем запись справочников через сырые строки 
#win_bonus = {777: 200, 999: 100, 555: 50, 333: 15, 111: 10, r'\d77': 5, r'\dd7': 3, r'\d00': 2, r'\dd0': 1}
#win_bonus['700']
# результат так себе. Ковырять буду позже. Пока сделаем в тупую, через генерацию справочников и объединение 
```

```python
#Создадим полный словарь выигрышных комбинаций
win_bonus = {'777': 200, '999': 100, '555': 50, '333': 15, '111': 10}
x1 = {f'{i}'+'77':5 for i in range(0,10)}
x2 = {f'{i}'+'00':2 for i in range(0,10)}
x3 = {f'{i}'+'7':3 for i in range(10,100)}
x4 = {f'{i}'+'0':1 for i in range(10,100)}
win_bonus_dict = x4 | x3 | x2 | x1 | win_bonus
print(len(win_bonus_dict))
```

```python
trials = 1000000
win_value = 0
success = 0
for i in range(trials):
    y1 = np.random.randint(0,10)
    y2 = np.random.randint(0,10)
    y3 = np.random.randint(0,10)
    y = str(y1)+str(y2)+str(y3)
    if y in win_bonus_dict.keys():
        win_value += win_bonus_dict[y]
        success += 1 
print(f'Кол-во попыток: {trials} раз')
print(f'Кол-во выигрышей: {success} раз')
print(f'Затраченная сумма: {trials} рублей')
print(f'Сумма выигрыша: {win_value} рублей')
print('----------------------------------')
if win_value < trials : 
    print(f'Игра невыгодна. Размер потерь: {trials - win_value} рублей')
else:
    print(f'Игра выгодна. Общий выигрыш: {win_value - trials} рублей')
print('----------------------------------')    
print(f"""
Расчетная вероятность выигрыша составляет: {(len(win_bonus_dict.keys()) / 1000)*100}%
Частота выигрыша при моделировании составила {success / trials * 100}%
Смоделированный выигрыш(проигрыш): {round(((win_value - trials) / trials),2)} рублей.""" ) 
```

## Дополнительное задание повышенной сложности

### Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования
Используем формулу риска. Веротяность события на ущерб от этого события 

В нашем случае - это сумма рисков всех событий. Отдельные риски - вероятность каждого события (1/1000) * на сумму выигрыша. 

Для призовых комбинаций посчитаем значения через справочник, а для проигрышей возьмем вероятность проигрыша. 

```python
loss_risk = ((1000 - len(win_bonus_dict.keys())) / 1000)*(-1)
win_risk = 0
for win_count in win_bonus_dict.values():
    win_risk += (win_count-1)/1000
avr_value = round(win_risk + loss_risk, 2)

print(f"""Теоритический средний выигрыш(проигрыш) составит: {avr_value} рублей 
Смоделированный выигрыш(проигрыш): {round(((win_value - trials) / trials),2)} рублей.""" ) 
```

Средний выигрыш **теоритический и смоделированный практически равны** на 1 000 000 попыток. На 100 млн попыток значения приближаются вплоть до 4-го знака.
